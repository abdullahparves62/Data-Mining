```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder()
```

```python
covid=covid.dropna()
```

```python
result = onehot.fit_transform(covid[['State/UTs']])
```

```python
print(result)
```

```python
newdata = covid.join(pd.DataFrame(result.toarray(),columns=onehot.categories_))
```
